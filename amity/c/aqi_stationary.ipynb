```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
import datetime

# Load dataset
df = pd.read_csv(r'C:\Users\ASUS\Desktop\Climate1\AQI.csv')

# Convert date column to datetime format
df["date"] = pd.to_datetime(df["date"])

# Sort by date
df = df.sort_values(by="date")

# Get unique cities
cities = df["City"].unique()

# Function to check stationarity
def check_stationarity(city):
    city_data = df[df["City"] == city].groupby("date")["Index Value"].mean()

    # Perform Augmented Dickey-Fuller (ADF) test
    adf_test = adfuller(city_data)

    # Plot time-series data
    plt.figure(figsize=(10, 5))
    plt.plot(city_data, label=f"AQI Index Value - {city}", color="blue")
    plt.title(f"Time Series of AQI Index Value ({city})")
    plt.xlabel("Date")
    plt.ylabel("AQI Value")
    plt.legend()
    plt.grid()
    plt.show()

    # Print ADF Test results
    print(f"📊 **ADF Test for {city}**")
    print(f"   - Test Statistic: {adf_test[0]}")
    print(f"   - p-value: {adf_test[1]}")
    print(f"   - Conclusion: {'✅ Stationary' if adf_test[1] < 0.05 else '❌ Non-Stationary'}")
    print("-" * 50)

# Run test for each city
for city in cities:
    check_stationarity(city)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import joblib
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler
```

```python
def load_data(file_path):
    df = pd.read_csv(file_path)
    df["date"] = pd.to_datetime(df["date"])
    df = df.sort_values(by="date")
    return df
```

```python
def check_stationarity(city, df):
    city_data = df[df["City"] == city].groupby("date")["Index Value"].mean()
    adf_test = adfuller(city_data.dropna())
    return adf_test[1] < 0.05
```

```python
def fit_arima(city, df):
    city_data = df[df["City"] == city].groupby("date")["Index Value"].mean()
    model = ARIMA(city_data, order=(1, 1, 1))
    model_fit = model.fit()
    joblib.dump(model_fit, f"arima_{city}.pkl")
    print(f"ARIMA Model for {city} saved successfully.")
```

```python
def fit_lstm(city, df):
    city_data = df[df["City"] == city].groupby("date")["Index Value"].mean().dropna()
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(city_data.values.reshape(-1, 1))
    joblib.dump(scaler, f"scaler_{city}.pkl")

    X, y = [], []
    for i in range(1, len(scaled_data)):
        X.append(scaled_data[i-1:i])
        y.append(scaled_data[i])

    X, y = np.array(X), np.array(y)

    model = Sequential()
    model.add(LSTM(units=50, return_sequences=True, input_shape=(X.shape[1], X.shape[2])))
    model.add(LSTM(units=50))
    model.add(Dense(units=1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    model.fit(X, y, epochs=10, batch_size=32, verbose=0)
    model.save(f"lstm_{city}.h5")
    print(f"LSTM Model for {city} saved successfully.")
```

```python
def load_arima(city):
    return joblib.load(f"arima_{city}.pkl")

def load_lstm(city):
    return load_model(f"lstm_{city}.h5"), joblib.load(f"scaler_{city}.pkl")
```

```python
file_path = r'C:\Users\ASUS\Desktop\Climate1\AQI.csv'  # Update with your file path
df = load_data(file_path)
cities = df["City"].unique()

stationary_cities = [city for city in cities if check_stationarity(city, df)]
non_stationary_cities = [city for city in cities if city not in stationary_cities]

for city in stationary_cities:
    fit_arima(city, df)

for city in non_stationary_cities:
    fit_lstm(city, df)

print("✅ Mixed approach applied successfully! Models saved for future use.")
```

```python
def predict_aqi(city, future_date):
    # Ensure future_date is a pandas Timestamp
    if isinstance(future_date, str):
        try:
            future_date = pd.to_datetime(future_date, format="%Y-%m-%d")
        except Exception as e:
            return f"Invalid date format! Error: {e}"

    # Get the last available date from dataset
    last_date = df["date"].max()
    if not isinstance(last_date, pd.Timestamp):
        last_date = pd.to_datetime(last_date)

    # Compute days ahead
    days_ahead = (future_date - last_date).days

    if days_ahead <= 0:
        return "Future date should be beyond the available data."

    # ARIMA for stationary cities
    if city in stationary_cities:
        model = load_arima(city)
        forecast = model.forecast(steps=days_ahead)
        return forecast.iloc[-1]  # Ensure index-safe access

    # LSTM for non-stationary cities
    elif city in non_stationary_cities:
        model, scaler = load_lstm(city)
        last_value = df[df["City"] == city]["Index Value"].values[-1]
        last_scaled = scaler.transform(np.array([[last_value]]))

        for _ in range(days_ahead):
            input_data = np.array(last_scaled[-1:]).reshape(1, 1, 1)
            pred = model.predict(input_data)
            last_scaled = np.append(last_scaled, pred)

        predicted_aqi = scaler.inverse_transform(np.array([last_scaled[-1]]).reshape(-1, 1))
        return predicted_aqi[0][0]

    else:
        return "City not found in dataset!"
```

```python
from prophet import Prophet
```

```python
def fit_prophet(city, df):
    city_data = df[df["City"] == city].groupby("date")["Index Value"].mean().reset_index()
    city_data.rename(columns={"date": "ds", "Index Value": "y"}, inplace=True)

    model = Prophet()
    model.fit(city_data)

    joblib.dump(model, f"prophet_{city}.pkl")
    print(f"Prophet Model for {city} saved successfully.")
```

```python
def predict_arima(Agra, future_steps=7):
    model = load_arima(Agra)
    forecast = model.forecast(steps=future_steps)
    return forecast
```
